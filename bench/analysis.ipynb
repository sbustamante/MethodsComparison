# General Functions

```python
#Fold where complete data are stored
datafold = '/home/sebastian/data/Simulations'
#Fold where local data are stored
localdatafold = './data/'
#Codes Fold
codesfold = "./codes/"

#==================================================================================================
#Parameters
#==================================================================================================
rhomean = 0.272*277.458477853 # M_sun/kpc^3 #MATTER
rhomean = 0.046*278.458477853 # M_sun/kpc^3 #BARYONS

#==================================================================================================
#Used functions
#==================================================================================================
# Saving data in ascii format
def ascii( snapbase, snap, files, sampling, type ):
    #Running code
    os.system( "%s/Ascii.out %s_%03d %d %d temp.tmp %d"%\
    (codesfold, snapbase, snap, files, sampling, type) )
    data = np.loadtxt( "temp.tmp" )
    os.system( "rm temp.tmp" )
    return data

# Cutting the simulation box
def slide( snapbase, snap, files, axis, coor, dx, sampling, type ):
    #Running code
    os.system( "%s/Cutter.out %s_%03d %d %d %f %f %d temp.tmp %d"%\
    (codesfold, snapbase, snap, files, axis, coor, dx, sampling, type) )
    data = np.loadtxt( "temp.tmp" )
    os.system( "rm temp.tmp" )
    return data

# Finding distance to nearest halo in units of virial radius
def gasdomains( X, Y, Z, sim, snapbase, snap, Lbox ):
    #Loading indexes
    index = np.loadtxt( "%s/%s/%s_%d.domain"%(localdatafold, sim, snapbase, snap) )
    #Loading catalogue of halos
    halos = np.loadtxt( "%s/%s/%s_%d.halo_catalog"%(localdatafold, sim, snapbase, snap) )
    
    #Calculating distances normalized with virial radius and virial mass
    Ldomain = []
    Mdomain = []
    for i in xrange( 0, len(X) ):
        #Position of domainant halo
        Xh = halos[index[i],0]; Yh = halos[index[i],1]; Zh = halos[index[i],2]
        #Virial radius
        Rvir = halos[index[i],4]
        #Virial mass
        Mdomain.append(halos[index[i],4])
        #Distance
        Xrel = min( abs(X[i]-Xh), abs(X[i]-Xh+Lbox), abs(X[i]-Xh-Lbox) )
        Yrel = min( abs(Y[i]-Yh), abs(Y[i]-Yh+Lbox), abs(X[i]-Yh-Lbox) )
        Zrel = min( abs(Z[i]-Zh), abs(Z[i]-Zh+Lbox), abs(Z[i]-Zh-Lbox) )
        #Relative distance
        Ldomain.append( norm([Xrel,Yrel,Zrel])/Rvir )
	
    Ldomain = np.array(Ldomain)
    Mdomain = np.array(Mdomain)
    
    return Ldomain, Mdomain
```

## Evolution

Next, it is shown the evolution of all the particles (DM+GAS) within a thin slide of the simulation **SPH_064**

```python
#==================================================================================================
#Animations of evolution of a cosmological simulation
#==================================================================================================
#......................................
#    PARAMETERS
#......................................
#Simulation
simulation = "SPH_064"
#Box lenght [kpc h^-1]
Box = 20000
#Snapbase
snapbase = "snap"
#Number of snapshots
snaps = 136
#Number of files per snap
snapfiles = 1
#Sampling rate
sampling = 1
#Type of print [ 0-all  1-gas   2-DM ]
typed = 0

#Axis     [0-X  1-Y  2-Z]
axis = 0
#Slide    [kpc h^-1]
coor = 10000
#Thick	[kpc h^-1]
dx = 3000

#......................................
#    ANIMATION
#......................................
#Initializing Figure
fig = plt.figure( figsize=(5,5) )
#Subplot for Voids
ax = fig.add_subplot(111, projection='3d')
ax = fig.add_subplot(111)
part_scatter, = ax.plot([], [], '.', color="white", markersize=0.2)
ax.set_xlim( (0, Box) )
ax.set_ylim( (0, Box) )
ax.fill_between( [0,Box], [0,0], [Box,Box], color = "black" )
ax.set_axis_off()

#Snap array
snaps_array = np.arange(snaps)

def init():
    part_scatter.set_data([], [])
    return part_scatter,

def animate(i):
    #Loading particle data
    particles = slide( "%s/%s/%s"%(datafold,simulation,snapbase),\
    snaps_array[i], snapfiles, axis, coor, dx, sampling, typed )
    #Setting new data
    part_scatter.set_data( particles[:,1], particles[:,2] )

    return part_scatter,

animation.FuncAnimation(fig, animate, init_func=init,
                        frames=snaps-1, interval=100, blit=True)
```

```python
#Energy of the simulation
def Energy( simulations ):
    #Detecting several simulations at the same time
    if type(simulations)!=str:
        nsim = len(simulations)
    else:
        simulations = np.array([simulations,])
        nsim = 1
        
    plt.figure(figsize=(7*(2 if nsim>1 else 1),5*np.ceil(nsim/2.0)))
        
    i_sim = 0
    for simulation in simulations:
        i_sim += 1
        #Loading energy
        energy = np.loadtxt( "%s/%s/energy.txt"%(datafold,simulation) )

        #Plotting energies
        if nsim>1:
            plt.subplot( int(nsim/2.0), (2 if nsim>1 else 1), i_sim )
        plt.plot( energy[:,0], energy[:,1], lw = 3, color="green", label="Internal energy" )
        plt.plot( energy[:,0], energy[:,2], lw = 3, color="red", label="Potential energy" )
        plt.plot( energy[:,0], energy[:,3], lw = 3, color="blue", label="Kinetic energy" )
        plt.plot( energy[:,0], energy[:,1]+energy[:,2]+energy[:,3], lw = 3.5, color="black", label="Total energy" )
        #format
        plt.grid(1)
        plt.legend( loc = "center right" )
        plt.title("Energy %s"%(simulation))
        plt.ylabel( "Energy" )
        plt.xlabel( "Time [$H_0^{-1}$]" )
```

## Domains of dark matter halos

[Contents](#Contents)

Next, it is shown a 3D snapshot at z=0 for all the particles (DM+GAS) for two simulations (SPH_064 and VPH_064). Dark matter halos are labeled using different colors. Domains of gas are also labeled using the same color of the associated halo.

```python
#==================================================================================================
#3D scatter of a snap of the simulation with colours indicating dark matter halos and gas domains
#==================================================================================================
def halos_3d( simulation, snap, gasFlag=True ):
    #......................................
    #    PARAMETERS
    #......................................
    #Box lenght [Mpc h^-1]
    Box = 20
    #Snapbase
    snapbase = simulation
    #Number of files per snap
    snapfiles = 1
    #Sampling rate
    sampling = 1
    #Type of print (0-all or 1-gas or 2-DM)
    type = 0
    
    #......................................
    #    PLOT
    #......................................
    fig = plt.figure( figsize = (20,10) )
    ax1 = fig.add_subplot(121, projection='3d')
    #Loading particle positions
    dm = ascii( "%s/%s/%s"%(datafold,simulation,snapbase), snap, snapfiles, sampling, 2 )
    gas = []
    if gasFlag:
        gas = ascii( "%s/%s/%s"%(datafold,simulation,snapbase), snap, snapfiles, sampling, 1 )
    
    #Loading indexes of host halos
    members = np.loadtxt( "%s/%s/%s_%d.members"%(localdatafold,simulation,snapbase,snap) )
    num_halos = np.max( members )
    colors = np.array([nan] + list(np.random.permutation( range(1,num_halos.astype(int)+1) )))
    #Random colors
    members = members[len(gas):len(gas)+len(dm)]
    members = colors[ members.astype(int) ]
    
    #Loading indexes of domains of halos
    domains = np.loadtxt( "%s/%s/%s_%d.domain"%(localdatafold,simulation,snapbase,snap) )
    #Random colors
    domains = domains[:len(gas)]
    domains = colors[ domains.astype(int)+1 ]

    #GAS
    if gasFlag:
        ax1.scatter( gas[:,1]/1000., gas[:,2]/1000., gas[:,3]/1000., c=domains, marker='.', s = 0.2, linewidth=0.01,
        cmap="spectral" )
        ax1.set_xlim( (0,Box) )
        ax1.set_ylim( (0,Box) )
        ax1.set_zlim( (0,Box) )
        ax1.set_xlabel( "Mpc h$^{-1}$" )
        ax1.set_ylabel( "Mpc h$^{-1}$" )
        ax1.set_zlabel( "Mpc h$^{-1}$" )
        ax1.set_title( "GAS in domains %s"%(simulation) )
    
    #DM
    ax2 = fig.add_subplot(122, projection='3d')
    ax2.scatter( dm[:,1]/1000., dm[:,2]/1000., dm[:,3]/1000., c=members, marker='.', s = 0.2, linewidth=0.01,
    cmap="spectral" )
    ax2.set_xlim( (0,Box) )
    ax2.set_ylim( (0,Box) )
    ax2.set_zlim( (0,Box) )
    ax2.set_xlabel( "Mpc h$^{-1}$" )
    ax2.set_ylabel( "Mpc h$^{-1}$" )
    ax2.set_zlabel( "Mpc h$^{-1}$" )
    ax2.set_title( "DM in halos %s"%(simulation) )
```

```python
#==================================================================================================
#Mass function of found halos
#==================================================================================================
def mass_function( simulation, snap ):
    #Snapbase
    snapbase = simulation
    
    mass = np.loadtxt( "%s/%s/%s_%d.halo_catalog"%(localdatafold,simulation,snapbase,snap), usecols=(3,) )*1e10
    #start with a rectangular Figure
    plt.figure(1, figsize=(10,6))
    #Making histogram
    numbers, mass = np.histogram( np.log10(mass), bins=np.sqrt(len(mass)) )
    plt.semilogy( mass[1:], numbers, lw = 2, label=simulation )
    plt.ylabel( "Number of halos" )
    plt.xlabel( "$\log_{10}$($M$  [h$^{-1}$M$_{\odot}$])" )
    plt.title("Mass function of halos")
    plt.grid(1)
    plt.legend()
    
```

# Phase Diagrams

Next, it is shown the phase diagram at z=0 for gas particles for two simulations (SPH_064 and VPH_064) 

```python
#==================================================================================================
#Phase diagram of a the gas particles for a given snapshot and for given simulation
#==================================================================================================
def PhaseDiagram( diagram, simulations, snap, video, Xlim = None, Ylim=None, Clim=None, bins2D = 300 ):
    """
    PhaseDiagram

    Function to plot phase diagrams of gas particles of simulation
    
    Arguments:
    -diagram
            <ABC>, The properties A and B are plotted in the x and y axis and the C property is 
            coloured. A, B or C may take one of the following options:
            U: energy
            R: density
            P: pressure
            T: temperatura
            M: mass
    -simulation
            Name of the simulation. (SPH, VPH)
    -snap
            Number of snaps (if video True), number of snap (if video False)
    -video
            Option to generate video
            
    """
    #......................................
    #    PARAMETERS
    #......................................
    #Box lenght [kpc h^-1]
    Box = 20000
    #Snapbase
    snapbase = "snap"
    #Number of snapshots
    snaps = snap
    #Number of files per snap
    snapfiles = 1
    #Sampling rate
    sampling = 1
    #Type of print (0-all or 1-gas or 2-DM)
    typep = 1
    
    #Gas properties
    prop_dict = {"U":"Energy", "R":"Density", "P":"Pressure", "T":"Temperature", "M":"Mass", "D":"Rdomain"}
    units = { \
    "Energy":		"(km/sec)$^2$",\
    "Density":		"-",\
    "Pressure":		"Pa",\
    "Temperature":	"K",\
    "Mass":			"$10^{10}$ h$^{-1}$M$_{\odot}$",\
    "Rdomain":		"R$_{vir}$"}
    norm = { \
    "Energy":		1,\
    "Density":		1e-10*rhomean,\
    "Pressure":		1,\
    "Temperature":	1,\
    "Mass":			1,\
    "Rdomain":		1}
    notat = { \
    "Energy":		"u",\
    "Density":		"$1+\delta$",\
    "Pressure":		"P",\
    "Temperature":	"T",\
    "Mass":			"M",\
    "Rdomain":		"R$_{dom}$"}
    ranges = { \
    "Energy":		[ 1, 6 ],\
    "Density":		[ -2, 7 ],\
    "Pressure":		[ -22, -11 ],\
    "Temperature":	[ 3, 8 ],\
    "Mass":			[ 1.6, 3.0 ],\
    "Rdomain":		[ -3.0, 2.5 ]}
    indexes = { "Energy":8, "Density":9, "Pressure":10, "Temperature":11, "Mass":15, "Rdomain":14 }
    properties = [prop_dict[prop] for prop in diagram]	#Specific order of properties
    
    #User defined ranges
    if Xlim != None:
        ranges[properties[0]] = Xlim
    if Ylim != None:
        ranges[properties[1]] = Ylim
    if Clim != None:
        ranges[properties[2]] = Clim
    
    #Window parameters
    prop1 = ranges[properties[0]]	#X
    prop2 = ranges[properties[1]]	#Y
    prop3 = ranges[properties[2]]	#Color
    
    prop1_hist = 5
    prop2_hist = 5
    #Bins of 1D histograms
    bins = 28*2.
    
    #Detecting several simulations at the same time
    if type(simulations)!=str:
        nsim = len(simulations)
    else:
        simulations = np.array([simulations,])
        nsim = 1
    
    #......................................
    #    MAKING VIDEO OF EVOLUTION
    #......................................
    if video == False:		#Plot
        snaprange = [snaps]
    else:				#Video
        snaprange = xrange( snaps )
            
    for snap in snaprange:
      i_sim = 0
      for simulation in simulations:
        snapbase = simulation
        #Formating plot....................................................
        #start with a rectangular Figure
        fig = plt.figure(1, figsize=(8,8))
        i_sim +=1
        #no labels
        nullfmt = NullFormatter()
    
        #definitions for the axes
        left, width = 0.1, 0.65
        bottom_v, height = 0.1, 0.65
        bottom_h = left_h = left+width+0.02
    
        rect_hist2D = [left, bottom_v, width, height]
        rect_histx = [left, bottom_h, 1.24*width, 0.2]
        rect_histy = [left_h, bottom_v, 0.2, height]
    
        axHist2D = fig.add_axes(rect_hist2D)
        axHistx = fig.add_axes(rect_histx)
        axHisty = fig.add_axes(rect_histy)
        
        #no labels
        axHistx.xaxis.set_major_formatter(nullfmt)
        axHisty.yaxis.set_major_formatter(nullfmt)
        
        #Loading data of current snap......................................
        data = ascii( "%s/%s/%s"%(datafold,simulation,snapbase), snap, snapfiles, sampling, typep )
        
        #Calculating domains...........................................................................
        for prop in diagram:
            if prop == "D" or prop == "M":
                L,M = gasdomains( data[:,1], data[:,2], data[:,3], simulation, snapbase, snap, Box )
                data2 = np.zeros( (len(data), len(data[0])+2 ) )
                data2[:,:-2] = data
                data2[:,-2] = L
                data2[:,-1] = M
                data = data2
                del data2
                break
        
        #Ordered properties
        property1 = np.log10(data[:,indexes[ properties[0] ]]/norm[ properties[0] ])
        property2 = np.log10(data[:,indexes[ properties[1] ]]/norm[ properties[1] ])
        property3 = data[:,indexes[ properties[2] ]]/norm[ properties[2] ]
        #Cutting properties according to range of property 3
        filt = (np.log10(property3)>=prop3[0])&(np.log10(property3)<=prop3[1])
        property1 = property1[ filt ]
        property2 = property2[ filt ]
        property3 = property3[ filt ]
          
        #Histogram2D.......................................................
        Hist_phase = np.transpose(np.histogram2d( property1, property2, weights = property3,
        bins = bins2D, range = (prop1, prop2) )[0][::,::-1] )
        Hist_phase_num = np.transpose(np.histogram2d( property1, property2,
        bins = bins2D, range = (prop1, prop2) )[0][::,::-1] )
        
        map2d = axHist2D.imshow( np.log10(Hist_phase[::,::]/Hist_phase_num[::,::]), \
        interpolation='nearest', aspect = 'auto', cmap = 'jet', \
        extent = (prop1[0],prop1[1],prop2[0],prop2[1]), vmin = prop3[0], vmax = prop3[1] )
        #Create the colorbar
        axc, kw = matplotlib.colorbar.make_axes( axHistx,\
        orientation = "vertical", shrink=1., pad=.04, aspect=10, anchor=(.5,1.5) )
        cb = matplotlib.colorbar.Colorbar( axc, map2d, orientation = "vertical" )
        cb.set_label( "log$_{10}$(%s  [%s])"%(notat[properties[2]],units[properties[2]]), \
        labelpad=-70, fontsize=12 )
        ticks = np.linspace( prop3[0],prop3[1],9 )
        cb.set_ticks( ticks )
        cb.set_ticklabels( ['{0:3.1f}'.format(t) for t in ticks] )
        
        #Histogram X (Prop1)
        histx = np.histogram( property1, bins=bins, normed=True, range=prop1 )
        axHistx.bar( histx[1][:-1], histx[0], width = (prop1[1]-prop1[0])/bins,\
        linewidth=2.0, color="gray" )
        #Histogram Y (Prop2)
        histy = np.histogram( property2, bins=bins, normed=True, range=prop2 )
        axHisty.barh( histy[1][:-1], histy[0], height = (prop2[1]-prop2[0])/bins,\
        linewidth=2.0, color="gray" )
        
        #Plot format.......................................................
        #Upper histogram
        axHistx.set_xlim( prop1 )
        axHistx.set_xticks( np.linspace( prop1[0],prop1[1],bins/4.-1 ) )
        yticks = np.linspace( 0, np.max(histx[0]), prop1_hist+1 )
        axHistx.set_yticks( yticks )
        axHistx.set_yticklabels( ['{0:1.2f}'.format(yt) for yt in yticks] )
        axHistx.grid( color='black', linestyle='--', linewidth=1., alpha=0.3 )
        axHistx.set_ylabel( "Normed distribution" )
        
        #Right histogram
        axHisty.set_ylim( prop2 )
        axHisty.set_yticks( np.linspace( prop2[0],prop2[1],bins/4.-1 ) )
        xticks = np.linspace( 0, np.max(histy[0]), prop2_hist+1 )
        axHisty.set_xticks( xticks )
        axHisty.set_xticklabels( ['{0:1.2f}'.format(xt) for xt in xticks], rotation=-90 )
        axHisty.grid( color='black', linestyle='--', linewidth=1., alpha=0.3 )
        axHisty.set_xlabel( "Normed distribution" )
        
        #2D Scatter
        #axHist2D.grid( color='w', linestyle='--', linewidth=1., alpha=0.3 )
        axHist2D.set_xlim( prop1 )
        axHist2D.set_ylim( prop2 )
        axHist2D.fill_between( prop1, [prop2[0], prop2[0]], [prop2[1], prop2[1]], color="black", zorder=0 )
        xticks = np.linspace( prop1[0],prop1[1],bins/4.-1 )
        axHist2D.set_xticks( xticks, ['{0:1.2f}'.format(xt) for xt in xticks] )
        yticks = np.linspace( prop2[0],prop2[1],bins/4.-1 )
        axHist2D.set_yticks( yticks, ['{0:1.2f}'.format(yt) for yt in yticks] )
        axHist2D.set_xlabel( "log$_{10}$(%s  [%s])"%(notat[properties[0]],units[properties[0]]) )
        axHist2D.set_ylabel( "log$_{10}$(%s  [%s])"%(notat[properties[1]],units[properties[1]]) )
        axHist2D.legend( loc='upper right', fancybox=True, shadow=True, ncol = 1, prop={'size':10} )
        axHist2D.text( prop1[1] - 0.98*(prop1[1]-prop1[0]),prop2[1] - 0.04*(prop2[1]-prop2[0]),\
        "%s:snapshot %03d (z = %1.2f)"%(simulation,snap,abs(data[0,12])), fontweight="bold", fontsize = 10, color='w' )
    
        if nsim>1:
            plt.savefig("sim%d.png"%(i_sim))
            plt.close()
    
      #Loading multiple images
      if nsim>1:
          fig = plt.figure(figsize=(8*(2 if nsim>1 else 1),8*np.ceil(nsim/2.0)))
          fig.subplots_adjust( left = 0.0, bottom = 0.0, right = 1.0, top = 1.0,
          wspace = 0.0, hspace = 0.0 )
          for i_sim in xrange(1, nsim+1):
                ax = fig.add_subplot( int(nsim/2.0), (2 if nsim>1 else 1), i_sim )
                img = mpimg.imread( "sim%d.png"%(i_sim) )
                ax.axis( "off" )
                ax.imshow( img, interpolation=None )
                os.system( "rm sim%d.png"%(i_sim) )
    
      #saving video images
      if video == True:		#Video
        fname='_tmp-%03d.png'%snap
        plt.savefig(fname)
        plt.close()
    
```
